```python
import pandas as pd

class Item:
    def __init__(self, name, qty, price):
        self.name = name
        self.qty = qty
        self.price = price
        self.free_shipping = False
    def total(self):
        return self.qty * self.price

cart = [
    Item("Widget", 12, 6.22),
    Item("Gadget", 5, 8.99),
    Item("Sprocket", 14, 10.20),
    Item("Bolt", 8, 6.64),
    Item("Washer", 12, 3.48),
    Item("Bracket", 12, 4.62),
    Item("Clip", 7, 12.85),
    Item("Pin", 9, 3.83),
    Item("Screw", 11, 6.77),
    Item("Nut", 15, 2.51),
    Item("Rivet", 14, 2.81),
    Item("Dowel", 2, 4.50),
    Item("Cap", 1, 14.22),
    Item("Plug", 7, 10.25),
    Item("Seal", 15, 5.03),
    Item("Ring", 8, 6.53),
]

rows = []
for item in cart:
    if item.qty >= 10:
        item.price *= 0.9
        rows.append({"type": "Price", "name": item.name, "value": f"{item.price:.2f}", "extra": f"qty={item.qty}"})
    if item.total() > 35:
        item.free_shipping = True
        rows.append({"type": "Free", "name": item.name, "value": str(item.free_shipping), "extra": f"total={item.total():.2f}"})

pd.DataFrame(rows)
```
